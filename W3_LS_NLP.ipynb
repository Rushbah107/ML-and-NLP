```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")     # Used to find Mood of the Person
print(classifier("I love Hugging Face!"))
```

```python
from datasets import load_dataset
dataset = load_dataset("imdb")      # Used this dataset to train the model for prediction of Hugging Face
print(dataset["train"][0])
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer("Hello, Hugging Face!", return_tensors="pt")
print(tokens)       # Print the tokenized of the text
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
model_name = "distilbert-base-uncased"
# AutoModelForSequenceClassification:- allows to load a pre-trained model for sequence classification
# AutoTokenizer:- converting raw text into numerical tokens that the model can process

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis", model=model_name)
print(classifier("Great tutorial!"))
```

```python
classifier = pipeline("text-classification")
print(classifier("Fun movie!"))
```

```python
def tokenize(examples):
  return tokenizer(examples["text"], padding="max_length", truncation=True)
tokenized_dataset = dataset.map(tokenize, batched=True)
```

```python
tokenized_dataset = tokenized_dataset.rename_column("label", "labels")
tokenized_dataset = tokenized_dataset.remove_columns(["text"])
tokenized_dataset.set_format("torch")       # This set ot yield PyTorch Tensor
```

```python
from transformers import Trainer, TrainingArguments
training_args = TrainingArguments(
    output_dir="./results",
    # evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=3,
)
trainer = Trainer(model=model, args=training_args, train_dataset=tokenized_dataset["train"])
trainer.train()
```

```python
model.save_pretrained("/content/drive/My Drive/models/fine_tuned_model")
```

## Explanation

In this machine learning pipeline, I fine-tune a pre-trained BERT-based model using Hugging Face’s transformers ('bert' used here) and datasets libraries to perform sentiment analysis on the IMDb dataset. The pipeline is loaded by the IMDb dataset with the datasets library, which provides easy access to standardized splits for training and evaluation. Next, the text data is tokenized using the tokenizer for bert-base-uncased, ensuring that inputs match the format the pre-trained model expects. The fine-tuning step leverages Trainer and TrainingArguments to set up training parameters such as batch size, learning rate, and evaluation strategy. After training, the model’s performance is evaluated using metrics like accuracy and F1-score to gauge how well it distinguishes positive and negative sentiment. Finally, the fine-tuned model and tokenizer are saved to Google Drive, enabling easy reuse or deployment. Overall, this pipeline demonstrates a modern and modular approach to NLP tasks by combining powerful pre-trained models with flexible fine-tuning and evaluation.
